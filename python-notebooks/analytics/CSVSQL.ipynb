```python
import pandas as pd
from pandasql import sqldf


# Load your CSV into a pandas DataFrame
df = pd.read_csv('../shared/DataLibrary/inputs/Crew/MNP/CSV/MNP_Clean.csv', low_memory=False)

query = """
SELECT SYS_CHANGE_OPERATION, SYS_CHANGE_VERSION, oh_OrderID
,oh_CampaignCode
,oh_CancelledDate
,oh_ClientCode
,oh_CreatedDate
,oh_CurrencyCode
,oh_CurrencyRate
,oh_CustomerID
,oh_DeliverByDate
,oh_DeliveryGrossValue
,oh_DeliveryNetValue
,oh_DeliveryReasonCode
,oh_DeliveryTaxCode
,oh_DeliveryTaxValue
,oh_DespatchDate
,oh_DiscountReasonCode
,oh_DiscountValue
,oh_DueDate
,oh_ExternalOrderID
,oh_LastActionCode
,oh_LoyaltyValue
,oh_MediaID
,oh_NetOrderValue
,oh_OrderGrossValue
,oh_OrderID
,oh_OrderMajorStatus
,oh_OrderMethod
,oh_OrderMinorStatus
,oh_OrderType
,oh_OrderValuePaid
,oh_OrderValueRefunded
,oh_OrderVoucherValue
,oh_PaymentMethod
,oh_PaymentType
,oh_PII
,oh_Priority
,oh_ShippingCode
,oh_ShippingReasonCode
,oh_SourceCode
,oh_TaxValue
,oh_UserID
FROM df AS header
GROUP BY SYS_CHANGE_OPERATION, SYS_CHANGE_VERSION, oh_OrderID
"""

# Execute the SQL query
result = sqldf(query, locals())

print(result)
```
